```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Let's assume your dataframe is loaded as df
# First, let's check for missing values
df=pd.read_csv('final_merged_all_1.csv')
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

# Display columns with missing values
missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent
})
print("Columns with missing values:")
print(missing_data[missing_data['Missing Values'] > 0].sort_values('Missing Values', ascending=False))

# Handle missing values
# For numerical columns, we'll fill with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].median())

# For categorical columns, we'll fill with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

# Convert date columns to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Extract additional date features if needed
df['day_of_week'] = df['order_date'].dt.dayofweek
df['day_of_month'] = df['order_date'].dt.day
df['quarter'] = df['order_date'].dt.quarter

# Check for and handle outliers
# Using IQR method for numerical columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap the outliers
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

# Normalize/scale relevant numerical features
# Selecting columns that might need scaling (price, investment, etc.)
cols_to_scale = ['gmv', 'list_price', 'product_mrp', 'Total Investment'] + [col for col in df.columns if 'SMA' in col or 'EMA' in col or 'Ad_Stock' in col]
scaler = StandardScaler()
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])

print("\nPreprocessing completed!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats



# Set display options for better viewing
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
plt.style.use('ggplot')

# Let's assume our preprocessed and encoded dataframe is df_encoded
# For better EDA, let's also keep a reference to original dataframe df

# 1. Basic statistics and overview
def basic_overview(df):
    print("Dataset Shape:", df.shape)
    print("\nData Types:")
    print(df.dtypes.value_counts())
    print("\nSample Data:")
    print(df.head())
    
    print("\nBasic Statistics for Numerical Features:")
    print(df.describe().T)
    
    # Memory usage
    memory_usage = df.memory_usage(deep=True).sum() / (1024 * 1024)  # in MB
    print(f"\nMemory Usage: {memory_usage:.2f} MB")

# 2. Time Series Analysis
def time_series_analysis(df):
    if 'order_date' not in df.columns:
        print("No order_date column found for time series analysis")
        return
    
    # Convert to datetime if not already
    df['order_date'] = pd.to_datetime(df['order_date'])
    
    # Group by date and calculate daily metrics
    daily_sales = df.groupby(df['order_date'].dt.date).agg({
        'gmv': 'sum',
        'units': 'sum',
        'order_id': pd.Series.nunique,
        'cust_id': pd.Series.nunique
    }).reset_index()
    daily_sales.columns = ['date', 'GMV', 'Units_Sold', 'Order_Count', 'Customer_Count']
    
    # Plot time series trends
    fig, axes = plt.subplots(2, 2, figsize=(18, 12))
    
    # GMV Trend
    axes[0, 0].plot(daily_sales['date'], daily_sales['GMV'])
    axes[0, 0].set_title('Daily GMV Trend')
    axes[0, 0].set_xlabel('Date')
    axes[0, 0].set_ylabel('GMV')
    axes[0, 0].tick_params(axis='x', rotation=45)
    
    # Units Sold Trend
    axes[0, 1].plot(daily_sales['date'], daily_sales['Units_Sold'])
    axes[0, 1].set_title('Daily Units Sold Trend')
    axes[0, 1].set_xlabel('Date')
    axes[0, 1].set_ylabel('Units Sold')
    axes[0, 1].tick_params(axis='x', rotation=45)
    
    # Order Count Trend
    axes[1, 0].plot(daily_sales['date'], daily_sales['Order_Count'])
    axes[1, 0].set_title('Daily Order Count Trend')
    axes[1, 0].set_xlabel('Date')
    axes[1, 0].set_ylabel('Number of Orders')
    axes[1, 0].tick_params(axis='x', rotation=45)
    
    # Customer Count Trend
    axes[1, 1].plot(daily_sales['date'], daily_sales['Customer_Count'])
    axes[1, 1].set_title('Daily Customer Count Trend')
    axes[1, 1].set_xlabel('Date')
    axes[1, 1].set_ylabel('Number of Customers')
    axes[1, 1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # Analyze by month, week, day of week
    if 'Month' in df.columns and 'Week' in df.columns:
        # Monthly trend
        monthly_sales = df.groupby('Month').agg({
            'gmv': 'sum',
            'units': 'sum'
        }).reset_index()
        
        fig, ax = plt.subplots(figsize=(12, 6))
        ax.bar(monthly_sales['Month'], monthly_sales['gmv'])
        ax.set_title('Monthly GMV Distribution')
        ax.set_xlabel('Month')
        ax.set_ylabel('GMV')
        plt.show()
        
        # Weekly trend
        weekly_sales = df.groupby('Week').agg({
            'gmv': 'sum',
            'units': 'sum'
        }).reset_index()
        
        fig, ax = plt.subplots(figsize=(14, 6))
        ax.bar(weekly_sales['Week'], weekly_sales['gmv'])
        ax.set_title('Weekly GMV Distribution')
        ax.set_xlabel('Week')
        ax.set_ylabel('GMV')
        plt.show()
    
    # Day of week analysis
    if 'day_of_week' in df.columns:
        day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        dow_sales = df.groupby('day_of_week').agg({
            'gmv': 'sum',
            'units': 'sum',
            'order_id': pd.Series.nunique
        }).reset_index()
        
        dow_sales['day_name'] = dow_sales['day_of_week'].apply(lambda x: day_names[x])
        
        fig, ax = plt.subplots(figsize=(12, 6))
        ax.bar(dow_sales['day_name'], dow_sales['gmv'])
        ax.set_title('GMV by Day of Week')
        ax.set_xlabel('Day of Week')
        ax.set_ylabel('GMV')
        plt.show()
    
    # Payday effect if available
    if 'payday_flag' in df.columns:
        payday_effect = df.groupby('payday_flag').agg({
            'gmv': ['sum', 'mean'],
            'units': ['sum', 'mean'],
            'order_id': pd.Series.nunique
        })
        
        print("\nPayday Effect Analysis:")
        print(payday_effect)
        
        # Visualize
        fig, ax = plt.subplots(figsize=(8, 6))
        payday_data = df.groupby('payday_flag')['gmv'].sum().reset_index()
        ax.bar(payday_data['payday_flag'].astype(str), payday_data['gmv'])
        ax.set_title('GMV by Payday Flag')
        ax.set_xlabel('Is Payday')
        ax.set_ylabel('GMV')
        plt.show()
    
    # Holiday effect if available
    if 'is_holiday' in df.columns:
        holiday_effect = df.groupby('is_holiday').agg({
            'gmv': ['sum', 'mean'],
            'units': ['sum', 'mean'],
            'order_id': pd.Series.nunique
        })
        
        print("\nHoliday Effect Analysis:")
        print(holiday_effect)
        
        # Visualize
        fig, ax = plt.subplots(figsize=(8, 6))
        holiday_data = df.groupby('is_holiday')['gmv'].sum().reset_index()
        ax.bar(holiday_data['is_holiday'].astype(str), holiday_data['gmv'])
        ax.set_title('GMV by Holiday Flag')
        ax.set_xlabel('Is Holiday')
        ax.set_ylabel('GMV')
        plt.show()

# 3. Category and Product Analysis
def category_analysis(df):
    # Check if relevant columns exist
    category_cols = [col for col in ['product_analytic_category', 'product_analytic_sub_category', 'product_analytic_vertical'] 
                    if col in df.columns]
    
    if not category_cols:
        print("No category columns found for analysis")
        return
    
    # For each category level, plot the top 10 by GMV
    for col in category_cols:
        # Top 10 categories by GMV
        top_categories = df.groupby(col).agg({
            'gmv': 'sum',
            'units': 'sum',
            'order_id': pd.Series.nunique
        }).sort_values('gmv', ascending=False).head(10).reset_index()
        
        # Plot
        fig, ax = plt.subplots(figsize=(14, 6))
        sns.barplot(x=col, y='gmv', data=top_categories, ax=ax)
        ax.set_title(f'Top 10 {col} by GMV')
        ax.set_xlabel(col)
        ax.set_ylabel('GMV')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
        
        # Category distribution (pie chart)
        fig, ax = plt.subplots(figsize=(10, 10))
        top_categories.plot.pie(y='gmv', labels=top_categories[col], autopct='%1.1f%%', ax=ax)
        ax.set_title(f'GMV Distribution by Top {col}')
        ax.set_ylabel('')
        plt.tight_layout()
        plt.show()

# 4. Marketing and Investment Analysis
def marketing_analysis(df):
    # Check if relevant columns exist
    investment_cols = [col for col in df.columns if any(x in col for x in ['Investment', 'TV', 'Digital', 'SEM', 'Radio', 'Sponsorship', 'Affiliates'])]
    
    if not investment_cols:
        print("No marketing investment columns found for analysis")
        return
    
    # Basic investment channel analysis
    base_channels = ['TV', 'Digital', 'Sponsorship', 'Content Marketing', 'Online marketing', 
                     'Affiliates', 'SEM', 'Radio', 'Other']
    available_channels = [ch for ch in base_channels if ch in df.columns]
    
    if available_channels:
        # Time series of marketing investments
        if 'order_date' in df.columns:
            # Group by date
            df['date'] = pd.to_datetime(df['order_date']).dt.date
            marketing_over_time = df.groupby('date')[available_channels].mean().reset_index()
            
            # Plot
            fig, ax = plt.subplots(figsize=(16, 8))
            for channel in available_channels:
                ax.plot(marketing_over_time['date'], marketing_over_time[channel], label=channel)
            
            ax.set_title('Marketing Investment by Channel Over Time')
            ax.set_xlabel('Date')
            ax.set_ylabel('Investment Amount')
            ax.legend()
            plt.tight_layout()
            plt.show()
        
        # Correlation between marketing channels and sales
        if 'gmv' in df.columns:
            correlation_data = df[available_channels + ['gmv', 'units']]
            correlation_matrix = correlation_data.corr()
            
            # Plot heatmap
            plt.figure(figsize=(12, 10))
            sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
            plt.title('Correlation Between Marketing Channels and Sales')
            plt.tight_layout()
            plt.show()
            
            # Marketing ROI analysis
            if 'Total Investment' in df.columns:
                # ROI calculation
                df['ROI'] = df['gmv'] / df['Total Investment'].where(df['Total Investment'] > 0, np.nan)
                
                # Aggregate by date
                roi_over_time = df.groupby('date').agg({
                    'gmv': 'sum',
                    'Total Investment': 'sum',
                    'ROI': 'mean'
                }).reset_index()
                
                # Plot
                fig, ax1 = plt.subplots(figsize=(16, 8))
                
                ax1.set_xlabel('Date')
                ax1.set_ylabel('Amount', color='tab:blue')
                ax1.plot(roi_over_time['date'], roi_over_time['gmv'], color='tab:blue', label='GMV')
                ax1.plot(roi_over_time['date'], roi_over_time['Total Investment'], color='tab:orange', label='Investment')
                ax1.tick_params(axis='y', labelcolor='tab:blue')
                ax1.legend(loc='upper left')
                
                ax2 = ax1.twinx()
                ax2.set_ylabel('ROI', color='tab:red')
                ax2.plot(roi_over_time['date'], roi_over_time['ROI'], color='tab:red', label='ROI')
                ax2.tick_params(axis='y', labelcolor='tab:red')
                
                plt.title('GMV, Investment and ROI Over Time')
                plt.tight_layout()
                plt.show()
            
            # Ad Stock Effect Analysis
            ad_stock_cols = [col for col in df.columns if 'Ad_Stock' in col]
            if ad_stock_cols:
                ad_stock_corr = df[ad_stock_cols + ['gmv']].corr()['gmv'].sort_values(ascending=False)[1:]
                
                plt.figure(figsize=(12, 6))
                ad_stock_corr.plot(kind='bar')
                plt.title('Correlation of Ad Stock Variables with GMV')
                plt.ylabel('Correlation Coefficient')
                plt.tight_layout()
                plt.show()

# 5. Customer Analysis
def customer_analysis(df):
    if 'cust_id' not in df.columns:
        print("No customer ID column found for analysis")
        return
    
    # Customer purchase frequency
    customer_orders = df.groupby('cust_id')['order_id'].nunique().reset_index()
    customer_orders.columns = ['cust_id', 'order_count']
    
    order_freq = customer_orders['order_count'].value_counts().sort_index().reset_index()
    order_freq.columns = ['orders_made', 'customer_count']
    
    plt.figure(figsize=(12, 6))
    plt.bar(order_freq['orders_made'].astype(str), order_freq['customer_count'])
    plt.title('Distribution of Customer Purchase Frequency')
    plt.xlabel('Number of Orders Made')
    plt.ylabel('Number of Customers')
    plt.tight_layout()
    plt.show()
    
    # Customer spend analysis
    customer_spend = df.groupby('cust_id')['gmv'].sum().reset_index()
    customer_spend.columns = ['cust_id', 'total_spend']
    
    # Histogram of customer spend
    plt.figure(figsize=(12, 6))
    plt.hist(customer_spend['total_spend'], bins=30)
    plt.title('Distribution of Customer Total Spend')
    plt.xlabel('Total Spend')
    plt.ylabel('Number of Customers')
    plt.tight_layout()
    plt.show()
    
    # Customer RFM analysis if we have date information
    if 'order_date' in df.columns:
        # Convert to datetime if not already
        df['order_date'] = pd.to_datetime(df['order_date'])
        
        # Get the most recent date in the dataset
        max_date = df['order_date'].max()
        
        # Calculate RFM
        rfm = df.groupby('cust_id').agg({
            'order_date': lambda x: (max_date - x.max()).days,  # Recency
            'order_id': pd.Series.nunique,                     # Frequency
            'gmv': 'sum'                                       # Monetary
        }).reset_index()
        
        rfm.columns = ['cust_id', 'recency', 'frequency', 'monetary']
        
        # Create RFM segments
        rfm['r_rank'] = pd.qcut(rfm['recency'], 5, labels=[5, 4, 3, 2, 1])
        rfm['f_rank'] = pd.qcut(rfm['frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
        rfm['m_rank'] = pd.qcut(rfm['monetary'], 5, labels=[1, 2, 3, 4, 5])
        
        # Calculate RFM score
        rfm['rfm_score'] = rfm['r_rank'].astype(int) + rfm['f_rank'].astype(int) + rfm['m_rank'].astype(int)
        
        # Create segments
        rfm['segment'] = pd.cut(rfm['rfm_score'], 
                               bins=[0, 5, 8, 11, 15], 
                               labels=['Hibernating', 'At Risk', 'Potential Loyalists', 'Champions'])
        
        # Plot RFM segments
        plt.figure(figsize=(10, 6))
        segment_counts = rfm['segment'].value_counts().sort_index()
        plt.bar(segment_counts.index, segment_counts.values)
        plt.title('Customer Segments by RFM Analysis')
        plt.xlabel('Segment')
        plt.ylabel('Number of Customers')
        plt.tight_layout()
        plt.show()
        
        # RFM heatmap
        rfm_agg = rfm.groupby('segment').agg({
            'recency': 'mean',
            'frequency': 'mean',
            'monetary': 'mean',
            'cust_id': 'count'
        }).reset_index()
        
        rfm_agg.columns = ['Segment', 'Recency (days)', 'Frequency', 'Monetary', 'Count']
        print("\nRFM Segment Summary:")
        print(rfm_agg)

# 6. Price and Discount Analysis
def price_discount_analysis(df):
    # Check if relevant columns exist
    if 'Discount%' not in df.columns or 'list_price' not in df.columns:
        print("Missing price or discount columns for analysis")
        return
    
    # Discount distribution
    plt.figure(figsize=(12, 6))
    plt.hist(df['Discount%'], bins=20)
    plt.title('Distribution of Discount Percentage')
    plt.xlabel('Discount %')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
    
    # Average GMV by discount range
    df['discount_range'] = pd.cut(df['Discount%'], 
                                 bins=[-0.1, 0, 10, 20, 30, 40, 50, 100], 
                                 labels=['0%', '1-10%', '11-20%', '21-30%', '31-40%', '41-50%', '51-100%'])
    
    discount_analysis = df.groupby('discount_range').agg({
        'gmv': ['mean', 'sum'],
        'units': ['mean', 'sum'],
        'order_id': pd.Series.nunique
    })
    
    print("\nDiscount Range Analysis:")
    print(discount_analysis)
    
    # Plot
    discount_gmv = df.groupby('discount_range')['gmv'].mean().reset_index()
    
    plt.figure(figsize=(12, 6))
    plt.bar(discount_gmv['discount_range'], discount_gmv['gmv'])
    plt.title('Average GMV by Discount Range')
    plt.xlabel('Discount Range')
    plt.ylabel('Average GMV')
    plt.tight_layout()
    plt.show()
    
    # Price elasticity (if we have enough data)
    # Simple approximation of price elasticity
    if 'product_analytic_category' in df.columns:
        # Group by category and discount range
        elasticity_data = df.groupby(['product_analytic_category', 'discount_range']).agg({
            'units': 'sum',
            'list_price': 'mean',
            'gmv': 'sum'
        }).reset_index()
        
        # Calculate percentage changes for key categories
        top_categories = df.groupby('product_analytic_category')['gmv'].sum().nlargest(5).index
        
        for category in top_categories:
            cat_data = elasticity_data[elasticity_data['product_analytic_category'] == category].sort_values('discount_range')
            
            if len(cat_data) > 1:
                plt.figure(figsize=(12, 6))
                plt.plot(cat_data['discount_range'], cat_data['units'], marker='o')
                plt.title(f'Units Sold vs Discount Range for {category}')
                plt.xlabel('Discount Range')
                plt.ylabel('Units Sold')
                plt.tight_layout()
                plt.show()

# 7. Delivery and SLA Analysis
def delivery_analysis(df):
    # Check if relevant columns exist
    delivery_cols = [col for col in ['deliverybdays', 'deliverycdays', 'sla', 'product_procurement_sla'] 
                    if col in df.columns]
    
    if not delivery_cols:
        print("No delivery or SLA columns found for analysis")
        return
    
    # Delivery days distribution
    for col in delivery_cols:
        if col in df.columns:
            plt.figure(figsize=(12, 6))
            df[col].hist(bins=20)
            plt.title(f'Distribution of {col}')
            plt.xlabel(col)
            plt.ylabel('Count')
            plt.tight_layout()
            plt.show()
    
    # Effect of delivery days on GMV
    if 'deliverybdays' in df.columns and 'gmv' in df.columns:
        df['delivery_bins'] = pd.cut(df['deliverybdays'], bins=[-1, 1, 2, 3, 5, 10, 100], 
                                    labels=['Same Day', '2 Days', '3 Days', '4-5 Days', '6-10 Days', '10+ Days'])
        
        delivery_gmv = df.groupby('delivery_bins').agg({
            'gmv': ['mean', 'sum'],
            'units': ['mean', 'sum'],
            'order_id': pd.Series.nunique
        })
        
        print("\nDelivery Time Impact Analysis:")
        print(delivery_gmv)
        
        # Plot
        plt.figure(figsize=(12, 6))
        delivery_plot = df.groupby('delivery_bins')['gmv'].mean().reset_index()
        plt.bar(delivery_plot['delivery_bins'], delivery_plot['gmv'])
        plt.title('Average GMV by Delivery Time')
        plt.xlabel('Delivery Time')
        plt.ylabel('Average GMV')
        plt.tight_layout()
        plt.show()

# 8. Payment Type Analysis
def payment_analysis(df):
    if 'order_payment_type' not in df.columns:
        print("No payment type column found for analysis")
        return
    
    # Payment type distribution
    payment_dist = df.groupby('order_payment_type').agg({
        'gmv': 'sum',
        'units': 'sum',
        'order_id': pd.Series.nunique
    }).reset_index()
    
    # Calculate percentages
    payment_dist['gmv_percent'] = payment_dist['gmv'] / payment_dist['gmv'].sum() * 100
    payment_dist['order_percent'] = payment_dist['order_id'] / payment_dist['order_id'].sum() * 100
    
    print("\nPayment Type Analysis:")
    print(payment_dist)
    
    # Plot
    plt.figure(figsize=(12, 6))
    plt.bar(payment_dist['order_payment_type'], payment_dist['gmv'])
    plt.title('GMV by Payment Type')
    plt.xlabel('Payment Type')
    plt.ylabel('GMV')
    plt.tight_layout()
    plt.show()
    
    # Pie chart
    plt.figure(figsize=(10, 10))
    plt.pie(payment_dist['gmv_percent'], labels=payment_dist['order_payment_type'], autopct='%1.1f%%')
    plt.title('GMV Distribution by Payment Type')
    plt.tight_layout()
    plt.show()

# 9. Weather and External Factors Analysis
def external_factors_analysis(df):
    # Check if relevant columns exist
    weather_cols = [col for col in ['Max Temp', 'Min Temp', 'Mean Temp', 'Total Rain (mm)', 'Total Snow (cm)'] 
                   if col in df.columns]
    
    if not weather_cols:
        print("No weather columns found for analysis")
        return
    
    # Correlation with sales
    if 'gmv' in df.columns:
        weather_corr = df[weather_cols + ['gmv', 'units']].corr()
        
        plt.figure(figsize=(12, 10))
        sns.heatmap(weather_corr, annot=True, cmap='coolwarm', center=0)
        plt.title('Correlation Between Weather Factors and Sales')
        plt.tight_layout()
        plt.show()
    
    # Temperature impact on sales
    if 'Mean Temp' in df.columns and 'gmv' in df.columns:
        # Create temperature bins
        df['temp_bins'] = pd.cut(df['Mean Temp'], bins=5)
        
        temp_gmv = df.groupby('temp_bins').agg({
            'gmv': ['mean', 'sum'],
            'units': ['mean', 'sum']
        })
        
        print("\nTemperature Impact Analysis:")
        print(temp_gmv)
        
        # Plot
        plt.figure(figsize=(12, 6))
        temp_plot = df.groupby('temp_bins')['gmv'].mean().reset_index()
        plt.bar(temp_plot['temp_bins'].astype(str), temp_plot['gmv'])
        plt.title('Average GMV by Temperature Range')
        plt.xlabel('Temperature Range')
        plt.ylabel('Average GMV')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    
    # Precipitation impact on sales
    if 'Total Precip (mm)' in df.columns and 'gmv' in df.columns:
        # Create precipitation bins
        df['precip_bins'] = pd.cut(df['Total Precip (mm)'], 
                                  bins=[-0.1, 0, 5, 10, 20, 100], 
                                  labels=['No Rain', 'Light Rain', 'Moderate Rain', 'Heavy Rain', 'Very Heavy Rain'])
        
        precip_gmv = df.groupby('precip_bins').agg({
            'gmv': ['mean', 'sum'],
            'units': ['mean', 'sum']
        })
        
        print("\nPrecipitation Impact Analysis:")
        print(precip_gmv)
        
        # Plot
        plt.figure(figsize=(12, 6))
        precip_plot = df.groupby('precip_bins')['gmv'].mean().reset_index()
        plt.bar(precip_plot['precip_bins'], precip_plot['gmv'])
        plt.title('Average GMV by Precipitation Level')
        plt.xlabel('Precipitation Level')
        plt.ylabel('Average GMV')
        plt.tight_layout()
        plt.show()

# 10. NPS Analysis
def nps_analysis(df):
    if 'NPS' not in df.columns:
        print("No NPS column found for analysis")
        return
    
    # NPS distribution
    plt.figure(figsize=(12, 6))
    plt.hist(df['NPS'], bins=11, range=(-0.5, 10.5))
    plt.title('Distribution of NPS Scores')
    plt.xlabel('NPS Score')
    plt.ylabel('Count')
    plt.xticks(range(0, 11))
    plt.tight_layout()
    plt.show()
    
    # Calculate NPS categories
    df['nps_category'] = pd.cut(df['NPS'], 
                              bins=[-0.1, 6, 8, 10], 
                              labels=['Detractor', 'Passive', 'Promoter'])
    
    nps_dist = df['nps_category'].value_counts(normalize=True) * 100
    print("\nNPS Category Distribution:")
    print(nps_dist)
    
    # Calculate overall NPS
    promoters = nps_dist.get('Promoter', 0)
    detractors = nps_dist.get('Detractor', 0)
    nps_score = promoters - detractors
    
    print(f"\nOverall NPS Score: {nps_score:.1f}")
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar(nps_dist.index, nps_dist.values)
    plt.title('NPS Category Distribution')
    plt.xlabel('NPS Category')
    plt.ylabel('Percentage (%)')
    plt.tight_layout()
    plt.show()
    
    # NPS impact on sales if we have time-series data
    if 'order_date' in df.columns:
        df['date'] = pd.to_datetime(df['order_date']).dt.date
        nps_time = df.groupby('date').agg({
            'NPS': 'mean',
            'gmv': 'sum',
            'units': 'sum'
        }).reset_index()
        
        # Plot
        fig, ax1 = plt.subplots(figsize=(16, 8))
        
        ax1.set_xlabel('Date')
        ax1.set_ylabel('NPS', color='tab:blue')
        ax1.plot(nps_time['date'], nps_time['NPS'], color='tab:blue')
        ax1.tick_params(axis='y', labelcolor='tab:blue')
        
        ax2 = ax1.twinx()
        ax2.set_ylabel('GMV', color='tab:red')
        ax2.plot(nps_time['date'], nps_time['gmv'], color='tab:red')
        ax2.tick_params(axis='y', labelcolor='tab:red')
        
        plt.title('NPS and GMV Over Time')
        plt.tight_layout()
        plt.show()
        
        # Correlation
        corr = nps_time[['NPS', 'gmv', 'units']].corr()
        print("\nCorrelation between NPS and Sales Metrics:")
        print(corr)

# 11. Stock Index Analysis
def stock_index_analysis(df):
    if 'Stock Index' not in df.columns:
        print("No Stock Index column found for analysis")
        return
    
    # Stock index distribution
    plt.figure(figsize=(12, 6))
    plt.hist(df['Stock Index'], bins=20)
    plt.title('Distribution of Stock Index')
    plt.xlabel('Stock Index')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
    
    # Stock index impact on sales
    if 'order_date' in df.columns:
        df['date'] = pd.to_datetime(df['order_date']).dt.date
        stock_time = df.groupby('date').agg({
            'Stock Index': 'mean',
            'gmv': 'sum',
            'units': 'sum'
        }).reset_index()
        
        # Plot
        fig, ax1 = plt.subplots(figsize=(16, 8))
        
        ax1.set_xlabel('Date')
        ax1.set_ylabel('Stock Index', color='tab:blue')
        ax1.plot(stock_time['date'], stock_time['Stock Index'], color='tab:blue')
        ax1.tick_params(axis='y', labelcolor='tab:blue')
        
        ax2 = ax1.twinx()
        ax2.set_ylabel('GMV', color='tab:red')
        ax2.plot(stock_time['date'], stock_time['gmv'], color='tab:red')
        ax2.tick_params(axis='y', labelcolor='tab:red')
        
        plt.title('Stock Index and GMV Over Time')
        plt.tight_layout()
        plt.show()
        
        # Correlation
        corr = stock_time[['Stock Index', 'gmv', 'units']].corr()
        print("\nCorrelation between Stock Index and Sales Metrics:")
        print(corr)

# 12. Correlation Analysis & Feature Importance
def correlation_analysis(df):
    # Select numerical columns for correlation analysis
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns
    
    # Exclude ID columns and date-derived columns for clarity
    exclude_cols = ['fsn_id', 'order_id', 'order_item_id', 'cust_id', 'Year', 'Month', 'Week', 'day_of_week', 'day_of_month', 'quarter']
    corr_cols = [col for col in num_cols if col not in exclude_cols]
    
    # If there are too many columns, select the most important ones
    if len(corr_cols) > 30:
        # Prioritize sales, marketing, and external factors
        priority_cols = ['gmv', 'units', 'list_price', 'Discount%', 'Sale']
        priority_cols += [col for col in corr_cols if 'TV' in col or 'Digital' in col or 'SEM' in col or 'Ad_Stock' in col]
        priority_cols += ['NPS', 'Mean Temp', 'Total Precip (mm)', 'Stock Index']
        corr_cols = list(set(priority_cols).intersection(set(corr_cols)))
        
        # Ensure we don't exceed a reasonable number of columns
        if len(corr_cols) > 30:
            # Focus on base metrics and not their derivatives
            corr_cols = [col for col in corr_cols if not any(x in col for x in ['SMA', 'EMA', 'Ad_Stock'])]
            corr_cols += ['TV_Ad_Stock', 'Digital_Ad_Stock', 'SEM_Ad_Stock']  # Just keep a few key ad stock metrics
    
    # Calculate correlation matrix
    corr_matrix = df[corr_cols].corr()
    
    # Visualize correlation matrix
    plt.figure(figsize=(18, 16))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix of Key Metrics')
    plt.tight_layout()
    plt.show()
    
    # Focus on correlations with GMV and units
    if 'gmv' in corr_cols and 'units' in corr_cols:
        # Correlations with GMV
        gmv_corr = corr_matrix['gmv'].sort_values(ascending=False)
        
        plt.figure(figsize=(14, 10))
        sns.barplot(x=gmv_corr.values, y=gmv_corr.index)
        plt.title('Features Correlation with GMV')
        plt.xlabel('Correlation Coefficient')
        plt.tight_layout()
        plt.show()
        
        # Correlations with units
        units_corr = corr_matrix['units'].sort_values(ascending=False)
        
        plt.figure(figsize=(14, 10))
        sns.barplot(x=units_corr.values, y=units_corr.index)
        plt.title('Features Correlation with Units Sold')
        plt.xlabel('Correlation Coefficient')
        plt.tight_layout()
        plt.show()
    
    # Feature importance using Random Forest (if we want to)
    # This is commented out as it requires additional setup and might be computationally intensive
    """
    from sklearn.ensemble import RandomForestRegressor
    from sklearn.model_selection import train_test_split
    
    # Prepare data for modeling
    X = df[corr_cols].drop(['gmv', 'units'], axis=1, errors='ignore')
    y_gmv = df['gmv']
    
    # Handle missing values for modeling
    X = X.fillna(0)
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y_gmv, test_size=0.2, random_state=42)
    
    # Train a simple Random Forest model
    model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
    model.fit(X_train, y_train)
    
    # Get feature importances
    importances = model.feature_importances_
    feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
    feature_importance = feature_importance.sort_values('Importance', ascending=False)
    
    # Plot feature importances
    plt.figure(figsize=(14, 10))
    sns.barplot(x='Importance', y='Feature', data=feature_importance)
    plt.title('Feature Importance for GMV Prediction')
    plt.tight_layout()
    plt.show()
    """



```
